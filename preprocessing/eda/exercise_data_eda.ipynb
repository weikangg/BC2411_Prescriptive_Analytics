```python
import pandas as pd
import matplotlib.pyplot as plt

# Enable inline plotting (if using Jupyter Notebook)
%matplotlib inline

# Load the final merged dataset.
# Adjust the file path as needed â€“ here, we assume it is in the current working directory.
df = pd.read_csv('../../data/final/processed_final_exercises.csv')

# Display the first few rows
df.head()
```

```python
# Cell 2: Basic summary statistics

# Number of unique exercise names
num_unique_exercises = df['exercise_name'].nunique()
print("Unique exercise names:", num_unique_exercises)

# Average calories burned per minute
avg_calories_burned_per_min = df['calories_burned_per_min'].mean()
print("Average calories burned per minute:", avg_calories_burned_per_min)

# Count the number of exercises by difficulty level
difficulty_counts = df['difficulty_level'].value_counts()
print("\nCounts by difficulty level:")
print(difficulty_counts)

# Count the number of exercises by workout location
location_counts = df['workout_location'].value_counts()
print("\nCounts by workout location:")
print(location_counts)

# Count the number of exercises by activity type
activity_counts = df['activity_type'].value_counts()
print("\nCounts by activity type:")
print(activity_counts)
```

```python
# Cell 3: Visualizations of categorical variables

# Bar chart for difficulty level counts
plt.figure(figsize=(8, 6))
difficulty_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Exercises by Difficulty Level')
plt.xlabel('Difficulty Level')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Bar chart for workout location counts
plt.figure(figsize=(8, 6))
location_counts.plot(kind='bar', color='coral')
plt.title('Number of Exercises by Workout Location')
plt.xlabel('Workout Location')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Bar chart for activity type counts
plt.figure(figsize=(8, 6))
activity_counts.plot(kind='bar', color='lightgreen')
plt.title('Number of Exercises by Activity Type')
plt.xlabel('Activity Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
