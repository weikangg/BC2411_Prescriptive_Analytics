```python
import pandas as pd
import matplotlib.pyplot as plt

# Enable inline plotting (if using Jupyter Notebook)
%matplotlib inline

# Load the final merged dataset.
# Adjust the file path as needed â€“ here, we assume it is in the current working directory.
df = pd.read_csv('../../data/final/processed_final_diets.csv')

# Display the first few rows
df.head()
```

```python
# Total number of recipes in the dataset
num_recipes = df.shape[0]
print("Total number of recipes:", num_recipes)

# Unique diet types and their counts
diet_types = df['diet_type'].unique()
print("\nUnique diet types:", diet_types)
print("Count per diet type:")
print(df['diet_type'].value_counts())

# Unique cuisine types and their counts
cuisine_types = df['cuisine_type'].unique()
print("\nUnique cuisine types:", cuisine_types)
print("Count per cuisine type:")
print(df['cuisine_type'].value_counts())
```

```python
numeric_cols = ['total_time_in_minutes', 'calories', 'fat', 'carbs', 'protein']
print("Summary statistics for numeric columns:")
print(df[numeric_cols].describe())
```

```python
# Bar chart for diet_type counts
plt.figure(figsize=(8, 6))
df['diet_type'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Count of Recipes per Diet Type')
plt.xlabel('Diet Type')
plt.ylabel('Number of Recipes')
plt.tight_layout()
plt.show()

# Bar chart for cuisine_type counts
plt.figure(figsize=(10, 6))
df['cuisine_type'].value_counts().plot(kind='bar', color='coral')
plt.title('Count of Recipes per Cuisine Type')
plt.xlabel('Cuisine Type')
plt.ylabel('Number of Recipes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
df[numeric_cols].hist(bins=20, layout=(2, 3), color='lightgreen', edgecolor='black')
plt.suptitle('Distribution of Numeric Columns')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
